### convert duration_ms to duration on sec 

```python
df_tracks.head(2)
```

```python
df_tracks['duration'] = df_tracks['duration_ms'].apply(lambda x:round(x/1000))
df_tracks.drop('duration_ms',inplace=True, axis=1)
```

```python
df_tracks.head()
```

```python
cor_df = df_tracks.drop(['key','mode','explicit'],axis=1).corr(method='pearson')
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(cor_df,annot = True,fmt = '.1g', vmin=-1, vmax=1, center=0, cmap='inferno', linewidth=1, linecolor='Black')
heatmap.set_title('correlation Heatmap Between varibale')
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
```

```python
print(len(sample_df))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y='loudness',x = 'energy', color='c').set(title='loudness vs energy correlations')
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y='popularity',x = 'acousticness', color='b').set(title='popularity vs acousticness')
```

```python
df_tracks['dates'] = df_tracks.index.get_level_values('release_date')
df_tracks.dates = pd.to_datetime(df_tracks.dates)
years = df_tracks.dates.dt.year
```

```python
df_tracks.head()
```

```python
sns.displot(years,discrete=True,aspect=2,height=5,kind='hist').set(title='number of songs per year')
```

```python
total_dr = df_tracks.duration
fig_dims = (18,7)
fix,ax = plt.subplots(figsize=fig_dims)
fig = sns.barplot(x=years,y=total_dr, ax=ax, errwidth = False).set(title='year Vs Durations')
plt.xticks(rotation=90)
plt.show()
```

```python
total_dr = df_tracks.duration
sns.set_style(style='whitegrid')
fig_dims = (10,5)
fig, ax = plt.subplots(figsize=fig_dims)
fig=sns.lineplot(x=years,y=total_dr, ax=ax,).set(title="year Vs Durations")
plt.xticks(rotation=60)
plt.show()
```

```python
plt.title('Duration of songs in diffrent genres')
sns.color_palette('rocket', as_cmap = True)
sns.barplot(y='genre', x='duration_ms', data=df_genre)
plt.xlabel('Duration in milli seconds')
plt.ylabel('genres')
```

```python
sns.set_style(style='darkgrid')
plt.figure(figsize=(10,5))
famous = df_genre.sort_values('popularity', ascending=False).head(10)
sns.barplot(y='genre',x='popularity', data=famous).set(title='top 5 genres by popularity')
```
